```python
Q1 = df['Total_salary'].quantile(0.25)
Q3 = df['Total_salary'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df['Total_salary'] = df['Total_salary'].apply(lambda x: max(lower_bound, min(upper_bound, x)))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(df['Total_salary'])
plt.title("Box Plot of Total Salary (After Outlier Treatment)")
plt.ylabel("Total Salary")
plt.show()
```

```python
def define_car_type(row):
    if row['Price'] >= 40000:
        return 'SUV'
    elif row['Price'] >= 25000:
        return 'Sedan'
    else:
        return 'Hatchback'
```
